```python
import pandas as pd
from copulas.multivariate import GaussianMultivariate
```

```python
# Path to the dataset
file_path = "/content/drive/MyDrive/loan_dataset_20000.csv"

# Load original dataset
df = pd.read_csv(file_path)

# Define Target
target = "loan_paid_back"

# Split into features and Target
X = df.drop(columns=[target])
y = df[target]

# Separate numeric and categorical columns
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns
```

```python
# Fit copula to numeric columns
copula = GaussianMultivariate()
copula.fit(df[num_cols])

# Sample new rows
synthetic_num = copula.sample(80000)
```

```python
# Resample categorical features
synthetic_cat = pd.DataFrame()

for col in cat_cols:
    probs = df[col].value_counts(normalize=True)
    synthetic_cat[col] = probs.sample(n=80000, replace=True).index
```

```python
# Combine numeric and categorical
synthetic = pd.concat([synthetic_num.reset_index(drop=True),
                       synthetic_cat.reset_index(drop=True)], axis=1)

# Combine synthetic data to original dataframe
expanded_df = pd.concat([df, synthetic], ignore_index=True)

print(expanded_df.shape)
```
